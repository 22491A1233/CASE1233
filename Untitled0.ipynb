```python
import pandas as pd
import matplotlib.pyplot as cr
import numpy as np
pavan=pd.read_csv('energy.csv')
print(pavan)
srh=pavan[pavan['State']=='Andhra']
rcb=pavan[pavan['State']=='Karnataka']
fig,diag=cr.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='yellow')
diag[1].plot(rcb['Date'],rcb['Consumption'],color='blue')
fig,diag=cr.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='yellow')
diag[1].bar(rcb['Date'],rcb['Consumption'],color='blue')
#pie chart
state1='Kerala'
state2='Delhi'
tdp=[state1,state2]
srh=pavan[pavan['State']=='Kerala']['Consumption'].sum()
rcb=pavan[pavan['State']=='Delhi']['Consumption'].sum()
ysr=[srh,rcb]
fig,diag=cr.subplots(2)
diag[0].pie(ysr,labels=tdp,autopct='%1.1f%%')
```
